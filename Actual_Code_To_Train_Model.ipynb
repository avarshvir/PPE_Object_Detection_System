```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8n.pt")
```

```python
model.train(
    data="data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    patience=10,
    device=0,
    project="PPE_Project",
    name="yolov8_ppe"
)
```

```python
model = YOLO("/content/runs/detect/PPE_Project/yolov8_ppe/weights/best.pt")

model.predict(
    source="test/images",
    conf=0.4,
    save=True
)
```

```python
metrics = model.val(
    data="data.yaml"
)

print("Precision:", metrics.box.mp)
print("Recall:", metrics.box.mr)
print("mAP@50:", metrics.box.map50)
print("mAP@50-95:", metrics.box.map)
```
